### 6. Basic String Compression

```python
s = "aaabbccccd"
compressed = ""
count = 1
for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        compressed += s[i-1] + str(count)
        count = 1
compressed += s[-1] + str(count)
print(compressed)
```

**Output:**

```
a3b2c4d1
```

### 14. Reverse a List In-place

### 15. Remove Duplicates While Preserving Order

```python
lst = [1, 2, 2, 3, 1, 4, 2]
seen = set()
unique = []
for item in lst:
    if item not in seen:
        seen.add(item)
        unique.append(item)
print(unique)
```

**Output:**

```
[1, 2, 3, 4]
```

### 17. Merge Two Sorted Lists

```python
a = [1, 3, 5]
b = [2, 4, 6]
merged = []
i = j = 0
while i < len(a) and j < len(b):
    if a[i] < b[j]:
        merged.append(a[i])
        i += 1
    else:
        merged.append(b[j])
        j += 1
merged.extend(a[i:])
merged.extend(b[j:])
print(merged)
```

**Output:**

```
[1, 2, 3, 4, 5, 6]
```

### **32. Merge two dictionaries, summing common values**

```python
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'b': 150, 'c': 100, 'd': 400}
merged = d1.copy()
for key, value in d2.items():
    merged[key] = merged.get(key, 0) + value
print("Merged Dictionary:", merged)
```

**Output:**

```
Merged Dictionary: {'a': 100, 'b': 350, 'c': 400, 'd': 400}
```

### **33. Access value in nested dictionary using list of keys**

```python
def get_nested_value(d, keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d

nested_dict = {'a': {'b': {'c': 42}}}
keys = ['a', 'b', 'c']
print("Value:", get_nested_value(nested_dict, keys))
```

**Output:**

```
Value: 42
```

### **35. Invert dictionary (values to keys; handle duplicates with lists)**

```python
def invert_dict(d):
    inverted = {}
    for k, v in d.items():
        if v not in inverted:
            inverted[v] = [k]
        else:
            inverted[v].append(k)
    return inverted

d = {'a': 1, 'b': 2, 'c': 1}
print("Inverted Dictionary:", invert_dict(d))
```

**Output:**

```
Inverted Dictionary: {1: ['a', 'c'], 2: ['b']}
```
